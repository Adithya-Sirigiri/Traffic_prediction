## Scatter plot

```python
from matplotlib import pyplot as plt
day=[1,2,3,4,5,6,7]
n=[4,6,2,7,6,2,6]

plt.scatter(day,n)
plt.show()
```

```python
from matplotlib import pyplot as plt
day=[1,2,3,4,5,6,7]
n=[4,6,2,7,6,2,6]
plt.xlabel("Day no in the week",fontsize=15)
plt.ylabel("No of sneezes",fontsize=15)
plt.title("Data of the patient",fontsize=15)
plt.scatter(day,n)
plt.show()
```

```python
from matplotlib import pyplot as plt
day=[1,2,3,4,5,6,7]
n=[4,6,2,7,6,2,6]
plt.xlabel("Day no in the week",fontsize=15)
plt.ylabel("No of sneezes",fontsize=15)
plt.title("Data of the patient",fontsize=15)
plt.scatter(day,n,c='r')
plt.show()
```

```python
from matplotlib import pyplot as plt
day=[1,2,3,4,5,6,7]
n=[4,6,2,7,6,2,6]
plt.xlabel("Day no in the week",fontsize=15)
plt.ylabel("No of sneezes",fontsize=15)
plt.title("Data of the patient",fontsize=15)
colors=['m','g','y','k','w','r','c']
size=[10,30,20,40,15,25,65]
plt.scatter(day,n,c=colors,s=size)
plt.show()
```

```python
from matplotlib import pyplot as plt
day=[1,2,3,4,5,6,7]
n=[4,6,2,7,6,2,6]
plt.xlabel("Day no in the week",fontsize=15)
plt.ylabel("No of sneezes",fontsize=15)
plt.title("Data of the patient",fontsize=15)
colors=[25,30,35,40,45,50,55]
size=[10,30,20,40,15,25,65]
a=plt.scatter(day,n,cmap='viridis',s=size)
z=plt.colorbar(a)
z.set_label='scale'
plt.show()
```
