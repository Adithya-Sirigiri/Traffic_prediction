## Bar Graph

```python
from matplotlib import pyplot as plt
x=["T.V","Fridge","Dryer","Remote"]
y=[45,60,55,75]

plt.bar(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=["T.V","Fridge","Dryer","Remote"]
y=[45,60,55,75]
plt.xlabel("Items",fontsize=15)
plt.ylabel("Qunatites",fontsize=15)
plt.title("Sales",fontsize=15)
plt.bar(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=["T.V","Fridge","Dryer","Remote"]
y=[45,60,55,75]
plt.xlabel("Items",fontsize=15)
plt.ylabel("Qunatites",fontsize=15)
plt.title("Sales",fontsize=15)
c=['r','g','b','m']
plt.bar(x,y,width=0.4,color=c)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=["T.V","Fridge","Dryer","Remote"]
y=[45,60,55,75]
plt.xlabel("Items",fontsize=15)
plt.ylabel("Qunatites",fontsize=15)
plt.title("Sales",fontsize=15)
c=['r','g','b','m']
plt.bar(x,y,width=0.4,color=c,align='center')
plt.show()
```

```python
from matplotlib import pyplot as plt
x=["T.V","Fridge","Dryer","Remote"]
y=[45,60,55,75]
plt.xlabel("Items",fontsize=15)
plt.ylabel("Qunatites",fontsize=15)
plt.title("Sales",fontsize=15)
c=['r','g','b','m']
plt.bar(x,y,width=0.4,color=c,align='center',edgecolor='black',linewidth=1.6)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=["T.V","Fridge","Dryer","Remote"]
y=[45,60,55,75]
plt.xlabel("Items",fontsize=15)
plt.ylabel("Qunatites",fontsize=15)
plt.title("Sales",fontsize=15)
c=['r','g','b','m']
plt.bar(x,y,width=0.4,color=c,align='center',edgecolor='black',linewidth=1.6,linestyle=':')
plt.show()
```

```python
from matplotlib import pyplot as plt
x=["T.V","Fridge","Dryer","Remote"]
y=[45,60,55,75]
plt.xlabel("Items",fontsize=15)
plt.ylabel("Qunatites",fontsize=15)
plt.title("Sales",fontsize=15)
c=['r','g','b','m']
plt.bar(x,y,width=0.4,color=c,align='center',edgecolor='black',linewidth=1.6,linestyle=':',label="Quanties sold")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
x=["T.V","Fridge","Dryer","Remote"]
y=[45,60,55,75]
plt.xlabel("Items",fontsize=15)
plt.ylabel("Qunatites",fontsize=15)
plt.title("Sales",fontsize=15)
c=['r','g','b','m']
plt.bar(x,y,width=0.4,color=c,align='center',edgecolor='black',linewidth=1.6,linestyle=':',label="Quanties sold",alpha=0.6)
plt.legend()
plt.show()
```
